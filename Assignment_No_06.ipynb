```python
import spacy
from spacy import displacy
nlp = spacy.load("en_core_web_sm")
```

```python
texts = ["Emily practices yoga every morning", "Alex enjoys painting landscapes", "Mark is writing a novel in his free time"]

for text in texts:
    doc = nlp(text)

    print(f"Dependency parse visualization: '{text}'")
    for token in doc:
        print(f'''
        TOKEN: {token.text}
                =====
                {token.tag_ = }
                {token.head.text = }
                {token.dep_ = }''')
```

```python
displacy.serve(doc, style="dep")
```
